```python
from multiprocessing import Pool
from newspaper import Article
import pandas as pd

def newspaper(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text
    except:
        pass

if __name__ == '__main__':
    header = pd.read_csv('CSV.header.fieldids.csv')
    header_list=list()
    url_list=list()
    for c in header:
        header_list.append(c)
    data_csv = pd.read_csv("final_gdelt.csv",names=header_list)
    data_size = len(data_csv)
    for count in range(data_size):
        url_list.append(data_csv["SOURCEURL"][count])
    p = Pool(3)
    result = p.map(newspaper, url_list)
    result = ['tidak ada berita' if x is None else x for x in result]
    news_data = [x.replace("\n","") for x in result]
    df = pd.DataFrame(news_data,columns=["berita"])
    df.to_csv('gdelt_dataset.csv', index=True, index_label="index")
    p.terminate()
    p.join()
```
